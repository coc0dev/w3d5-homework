```python
import requests
from IPython.display import clear_output
import pprint
import pandas as pd

class Pokemon:
    poke_list = []
#     df = pd.DataFrame(columns=['pokemon'])
    
    def add(self, get_pokemon):
#         takes data from API and creates a new character slot
        requesturl = "https://pokeapi.co/api/v2/pokemon/" + get_pokemon
        res = requests.get(requesturl)
        data = res.json()

        poke_dict = {
                    "types": [t['type']['name'] for t in data['types']],
                    "name": data['name'].title(),
                    "id": data['id'],
                    "height": data['height'],
                    "weight": data['weight'],
                    "stats": [(d['stat']['name'], d['base_stat']) for d in data['stats']],
                    "abilities": [ability['ability']['name'] for ability in data['abilities']]
                }
#         self.df.append({'poke_dict': poke_dict}, ignore_index=True)

        self.poke_list.append(poke_dict)
        print(f"\nYou added " + get_pokemon.title())
        
    
    def view(self, df, poke_list=''):
        # show all stored characters
        print(pprint.pprint(self.poke_list, sort_dicts = False))
#         print(self.df)
        
    def search(self, search_pokemon):
        requesturl = "https://pokeapi.co/api/v2/pokemon/" + search_pokemon
        res = requests.get(requesturl)
        data = res.json()

        poke_dict = {
                    "type": [t['type']['name'] for t in data['types']],
                    "name": data['name'].title(),
                    "id": data['id'],
                    "height": data['height'],
                    "weight": data['weight'],
                    "stats": [(d['stat']['name'], d['base_stat']) for d in data['stats']],
                    "abilities": [ability['ability']['name'] for ability in data['abilities']]
                }
        print(pprint.pprint(poke_dict))

class Pokedex:
    
    @classmethod
    def run(self):
        pk = Pokemon()
        
        active = True
        while active:
            choice = input("what would you like to do? add, view, search, or quit: ").lower()
            if choice == 'add':
                get_pokemon = input("Which character do you want to add? ").lower()
                pk.add(get_pokemon)
                    
            elif choice == 'view':
                pk.view(pk.poke_list)
            elif choice == 'search':
                search_pokemon = input("what character are you searching for? ").lower()
                pk.search(search_pokemon)
            else:
                choice == 'quit'
                active = False
        
Pokedex.run()
```
